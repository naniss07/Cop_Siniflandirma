```python
# Veriyi okuma ve işleme adımında kullanılacak olan kütüphaneler
import cv2
import urllib
import itertools
import numpy as np
import pandas as pd
import seaborn as sns
import random, os, glob
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
from urllib.request import urlopen
from imutils import paths


# Uyarıları kapatmak için kullanılacak kütüphaneler
import warnings
warnings.filterwarnings("ignore")

# Modeli değerlendirmek için kullanılacak kütüphaneler
from sklearn.metrics import classification_report, confusion_matrix

# Model değerlendirme için kullanılacak olan kütüphaneler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing import image
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from tensorflow.keras.layers import Conv2D, Flatten, MaxPooling2D, Dense, Dropout, SpatialDropout2D
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img, array_to_img

from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
```

```python
# Veri yolu ve etiketlerin tanımlanması
data_path = "Garbage classification"
labels_map = {"cardboard": 0, "glass": 1, "metal": 2, "paper": 3, "plastic": 4}
```

```python
# Veri setini okuyacak fonksiyonun tanımlanması

# Boyut bilgisi ve label değerlerinin belirlenmesi
target_size = (224, 224) # Normalde görüntüler 512x384 pikseldi ama biz görüntüleri 224x224 piksel yapacağız

labels_map = {"cardboard": 0, "glass": 1, "metal": 2, "paper": 3, "plastic": 4}
```

```python
import random

# Görselleri ve etiketleri saklayacağımız listeler
images = []
labels = []

# Her sınıf için maksimum alınacak resim sayısı
max_images_per_class = 400

# Alt klasörlerdeki görüntüleri işleme
for label_name, label_value in labels_map.items():
    folder_path = os.path.join(data_path, label_name)
    
    # Eğer klasör varsa işleme al
    if os.path.exists(folder_path):
        # Klasördeki tüm dosyaları al
        all_images = os.listdir(folder_path)
        
        # Resimlerin sırasını karıştır ve belirttiğiniz sayıda resim seç
        selected_images = random.sample(all_images, min(len(all_images), max_images_per_class))
        
        for img_name in selected_images:
            img_path = os.path.join(folder_path, img_name)
            
            # Görseli yükleyip yeniden boyutlandırmak
            img = cv2.imread(img_path)
            if img is not None:
                resized_img = cv2.resize(img, (224, 224))
                
                # Görselleri ve etiketleri listeye eklemek
                images.append(resized_img)
                labels.append(label_value)
    else:
        print(f"{label_name} klasörü bulunamadı, atlanıyor...")

# Numpy dizilerine dönüştürme
images = np.array(images)
labels = np.array(labels)

print("İşlenen görüntü sayısı:", len(images))
print("İşlenen etiket sayısı:", len(labels))
```

```python
# Kategorilerdeki görüntü sayısını hesaplama
category_counts = {}
for label_name, label_value in labels_map.items():
    count = np.sum(labels == label_value)
    category_counts[label_name] = count
    print(f"{label_name}: {count} adet resim")

# Görüntü dağılımını görselleştirmek için
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.bar(category_counts.keys(), category_counts.values())
plt.title('Kategorilerdeki Resim Sayıları')
plt.xlabel('Kategoriler')
plt.ylabel('Resim Sayısı')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Örnek görselleri görüntüleme
reverse_labels_map = {v: k for k, v in labels_map.items()}  
plt.figure(figsize=(12, 8)) 
for i in range(3):     
    idx = np.random.randint(0, len(images))  # Filtrelenmiş resimlerden rastgele seç
    img = images[idx]
    label = reverse_labels_map[labels[idx]]  # Etiketin ismini al
          
    plt.subplot(2, 3, i+1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  
    plt.title(label)
    plt.axis('off') 
plt.show()
```

```python




# Eğitim veri seti için ImageDataGenerator tanımlama
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    vertical_flip=False,
    fill_mode='nearest',
    brightness_range=[0.8, 1.2],  # Parlaklık değişimi
    channel_shift_range=50.0,     # Renk kanallarında hafif değişim
    validation_split=0.2           # Validation oranını artır
)

# Test veri seti için sadece ölçeklendirme
test_datagen = ImageDataGenerator(rescale=1./255, validation_split=0.1)
```

```python
# Eğitim veri seti için generator
train_generator = train_datagen.flow_from_directory(
    directory=data_path,
    target_size=target_size,
    class_mode="categorical",
    subset="training",  # Eğitim için ayrılan kısım
)

# Test veri seti için generator
test_generator = test_datagen.flow_from_directory(
    directory=data_path,
    target_size=target_size,
    class_mode="categorical",
    subset="validation",  # Doğrulama için ayrılan kısım
)
```

```python
# Batch kontrolü
x_train_batch, y_train_batch = next(train_generator)
x_test_batch, y_test_batch = next(test_generator)

print(f"x_train_batch shape: {x_train_batch.shape}")
print(f"y_train_batch shape: {y_train_batch.shape}")
print(f"x_test_batch shape: {x_test_batch.shape}")
print(f"y_test_batch shape: {y_test_batch.shape}")

# Kategori bilgilerini göster
unique_labels, label_counts = np.unique(labels, return_counts=True)
print("\nKategori Bilgileri:")
for label, count in zip(unique_labels, label_counts):
    print(f"Kategori {label}: {count} adet görüntü")

print(f"\nToplam kategori sayısı: {len(unique_labels)}")
```

```python
from tensorflow.keras.applications import ResNet50V2, EfficientNetB0
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout
from tensorflow.keras.models import Model

# Transfer Learning ile Daha Güçlü Model
base_model = ResNet50V2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = False  # Başlangıçta temel katmanları dondur

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(256, activation='relu')(x)
x = Dropout(0.5)(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.3)(x)
output = Dense(5, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=output)
```

```python
from tensorflow.keras.optimizers import Adam

# Model compile'ını categorical_crossentropy olarak değiştirin
model.compile(
    optimizer=Adam(learning_rate=1e-4),
    loss='categorical_crossentropy', 
    metrics=[
        'accuracy', 
        tf.keras.metrics.Precision(), 
        tf.keras.metrics.Recall()
    ]
)
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

callbacks = [
    EarlyStopping(
        monitor='val_accuracy', 
        patience=30, 
        restore_best_weights=True
    ),
    ReduceLROnPlateau(
        monitor='val_loss', 
        factor=0.2, 
        patience=10, 
        min_lr=1e-6
    ),
    ModelCheckpoint(
        filepath='best_model1.keras', 
        monitor='val_accuracy', 
        save_best_only=True,
        mode='max'
    )
]
```

```python

class_weights = {0: 1., 1: 1., 2: 1., 3: 1., 4: 1.}  # Örnek ağırlıklar

history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // train_generator.batch_size,
    validation_data=test_generator,
    class_weight=class_weights,
    validation_steps=test_generator.samples // test_generator.batch_size,
    epochs=45,  # Daha fazla epoch
    callbacks=callbacks
)
```
